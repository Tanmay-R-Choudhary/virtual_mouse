```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import os
```

```python
for i in os.listdir('mouse_mode/'):
    print(i)
```

```python
file_df = pd.DataFrame({
    'file': os.listdir('mouse_mode/')
})
```

```python
file_df.head()
```

```python
file_df['status'] = file_df.file.str.split('_', expand=True).loc[:, 2]
```

```python
file_df.head()
```

```python
df_arr = []

for i in range(file_df.shape[0]):
    df = pd.read_csv('mouse_mode/' + file_df.iloc[i, 0], header=None)
    df['status'] = [file_df.iloc[i, 1] for _ in range(df.shape[0])]
    df_arr.append(df)

for df in df_arr:
    print(df.head())
```

```python
df = pd.concat(df_arr, axis=0)
```

```python
df.head()
```

```python
df.shape
```

```python
sns.countplot(x=df.status)
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.head()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['status'], axis=1).to_numpy()
Y = df.status.astype('category').cat.codes.to_numpy()
```

```python
X.shape
```

```python
Y.shape
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y)
```

```python
xtrain.shape, xtest.shape, ytrain.shape, ytest.shape
```

```python
sns.countplot(x=ytrain)
```

```python
sns.countplot(x=ytest)
```

```python
xtrain_resampled, ytrain_resampled = SMOTE().fit_resample(xtrain, ytrain)
```

```python
xtrain_resampled.shape, ytrain_resampled.shape
```

```python
sns.countplot(x=ytrain_resampled)
```

# Model

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import f1_score
```

```python
model = RandomForestClassifier()
```

```python
model.fit(xtrain_resampled, ytrain_resampled)
```

```python
pred = model.predict(xtest)
```

```python
f1_score(pred, ytest)
```

```python
sns.countplot(x=ytest[pred != ytest])
```
